#### The results should look similar to what is shown in [BBC results page](https://www.bbc.com/sport/formula1/2021/abu-dhabi-grand-prix/results) 

What columns we need from different tables?
1. race_year, race_name, race_date from races_df
2. circuit_name from circuits_df
3. driver_name, driver_nationality, driver_number from drivers_df
4. team_name from constructors_df
5. grid, fastest_lap_time, time, points from results_df

Order of the columns displayed should be **race_year, race_name, circuit_name, race_date, driver_nationality, driver_name, team_name, driver_number, grid, fastest_lap_time, time, points**

How to combine different tables using joins?
1. We need circuit_name from circuits_df but results_df don't has circuit_id. So to get circuit_id we combine races_df with circuits_df
2. Combine results_df with the above result along with drivers_df and constructors_df

```python
# Join the races and circuits tables using inner join
races_circuits_df = races_df.join(circuits_df, races_df.circuit_id == circuits_df.circuit_id, "inner") \
                                      .select(races_df.race_id, races_df.race_year, races_df.race_name, circuits_df.circuit_name, races_df.race_date)
```

```python
# join results dataframe with the resultant above
race_results_df = results_df.join(races_circuits_df, results_df.race_id == races_circuits_df.race_id, "inner") \
                            .join(drivers_df, results_df.driver_id == drivers_df.driver_id, "inner") \
                            .join(constructors_df, results_df.constructor_id == constructors_df.constructor_id, "inner") \
                            .select(results_df.race_id, races_circuits_df.race_year, races_circuits_df.race_name, 
                                    races_circuits_df.circuit_name, races_circuits_df.race_date, drivers_df.nationality, drivers_df.driver_name, constructors_df.team_name, drivers_df.driver_number, results_df.grid, results_df.fastest_lap_time, results_df.time, results_df.points, results_df.position, results_df.file_date)
```

```python
display(race_results_df)
```
